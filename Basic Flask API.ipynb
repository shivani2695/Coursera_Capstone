```
from flask_ngrok import run_with_ngrok
from flask import Flask
import app
from flask import request, render_template, redirect
```

```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
```

```
df = pd.read_csv("Balanced-Data.csv")
```

```
Feature = df[['ROADCOND','LIGHTCOND','WEATHER','UNDERINFL','dayofweek']]
X = Feature
X= preprocessing.StandardScaler().fit(X).transform(X)
y = df['SEVERITYCODE'].values
```

```
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=1)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```
from sklearn.tree import DecisionTreeClassifier
drugTree = DecisionTreeClassifier(criterion="entropy", max_depth = 9)
drugTree.fit(X_train,y_train)
predTree = drugTree.predict(X_test)
```

we have earlier checked in Data Modiling-1 notebook that Decision tree is the best fit logrithm in tearms of accuracy so we will use decision tree here
